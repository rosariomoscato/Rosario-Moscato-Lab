```python
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("/content/invoice.jpg", cv2.IMREAD_COLOR)
```

```python
plt.rcParams["figure.figsize"] = (30,30)
plt.imshow(im)

plt.show()
```

```python
# Simple image to string
print(pytesseract.image_to_string(Image.open("/content/invoice.jpg")))

```

```python
# Get verbose data including boxes, confidences, line and page numbers
print(pytesseract.image_to_data(Image.open("/content/invoice.jpg")))
```

```python
# Get a searchable PDF
pdf = pytesseract.image_to_pdf_or_hocr('/content/invoice.jpg', extension='pdf')
with open('test.pdf', 'w+b') as f:
    f.write(pdf) # pdf type is bytes by default
```
